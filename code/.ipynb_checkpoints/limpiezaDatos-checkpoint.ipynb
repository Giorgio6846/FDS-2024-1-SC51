# Importe de librerias

```python
import pandas as pd
import json
```

```python
#dataBC = Data before cleaning
dataBC = pd.read_csv('../data/GBvideos_cc50_202101.csv')
```

```python
dataBC.info()
```

```python
dataBC.head()
```

```python
dataBC.isna().any()
```

```python
dataAC = dataBC
dataAC["trending_date"] = pd.to_datetime(dataBC["trending_date"], format = '%y.%d.%m')
dataAC.head()
```

```python
dataAC["publish_time"] = pd.to_datetime(dataBC["publish_time"])
dataAC.head()
```

# Read JSON Category Id

```python
catID = pd.read_json("./../data/GB_category_id.json")
category = pd.json_normalize(data = catID["items"], meta = ["id"])
```

```python
catID.head()
```

```python
category.info()
```

```python
category = category.drop(
    ["kind", "etag", "snippet.channelId", "snippet.assignable"], axis=1
)
```

```python
category.head()
```

```python
category["id"] = category["id"].astype('int64')
```

```python
dataAC = pd.merge(dataAC, category, left_on="category_id", right_on="id")
dataAC.head()
```

# Modificacion de datos

```python
dataAC["PercentageLikes"]    = round( ( ( (dataAC["likes"]) *100) / (dataAC["likes"] + dataAC["dislikes"]) ),2 )
dataAC["PercentageDislikes"] = round( ( ( (dataAC["dislikes"]) *100) / (dataAC["likes"] + dataAC["dislikes"]) ),2 )
```

```python
dataAC.head()
```

# Renombre de capitulos

```python
dataAC.rename(columns={"snippet.title": "CategoryTitle"}, inplace=True)
```

## Eliminacion de tablas auxiliares

```python
dataAC = dataAC.drop(["category_id", "id"], axis = 1)
```

```python
dataAC.head()
```

```python
dataAC = dataAC.sort_values(["video_id", "trending_date"], ascending = False)
dataAC.reset_index(inplace=True, drop=True)
dataAC.head()
```

# Guardado de csv limpio

```python
dataAC.to_csv("./../data/CDGBvideos_cc50_202101.csv", index=False)
```
